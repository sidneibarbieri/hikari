```python
import json
import re
import ipaddress
from tqdm import tqdm

def find_unique_internal_ips(file_path):
    internal_ips = set()
    ip_pattern = re.compile(r'\b(?:\d{1,3}\.){3}\d{1,3}\b')  # Regex para identificar endereços IPv4

    # Abrir o arquivo e calcular o número de linhas para a barra de progresso
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Processar cada linha com uma barra de progresso
    for line in tqdm(lines, desc="Processing IPv4 IPs"):
        log = json.loads(line)
        # Procurar por IPs em todos os valores de string
        for value in log.values():
            if isinstance(value, str):
                found_ips = ip_pattern.findall(value)
                for ip_address in found_ips:
                    try:
                        ip = ipaddress.ip_address(ip_address)
                        # Filtrar endereços especiais e verificar se é um endereço IPv4 e se está em uma faixa de rede interna
                        if ip.version == 4 and ip.is_private and ip_address not in ["127.0.0.1", "0.0.0.0", "255.255.255.255", "169.254.169.254"]:
                            internal_ips.add(ip_address)
                    except ValueError:
                        # Ignorar caso o valor não seja um endereço IP válido
                        continue

    return internal_ips

# Caminho do arquivo JSON
file_path = './apt29/apt29.json'
unique_ips = find_unique_internal_ips(file_path)

# Imprimir os endereços IP únicos encontrados
for ip in unique_ips:
    print(ip)
```
